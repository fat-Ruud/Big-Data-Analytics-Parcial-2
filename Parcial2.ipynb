```python
import requests
from urllib.request import urlopen

from bs4 import BeautifulSoup

import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://pdba.georgetown.edu/Executive/Colombia/pres.html"
```

```python
page = requests.get(url)
print(page) # Success making the request
```

```python
page.text
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
table = soup.find_all('table')
df = pd.read_html(str(table))[0]
```

```python
print(df)
```

```python
df_prueba = df

print(df[['Años / Years', 'Presidente / President','Partido / Party']])
```

```python
df_filtered = df_prueba[['Años / Years', 'Presidente / President','Partido / Party']].copy()
print(df_filtered.to_string(index=False))
```
